```python
class CreditCard:
    def __init__(self, customer, bank, account, limit, balance=0) -> None:
        self._customer = customer
        self._bank = bank
        self._account = account
        self._limit = limit
        self._balance = balance
    
    def get_customer(self):
        return self._customer
    
    def get_bank(self):
        return self._bank
    
    def get_account(self):
        return self._account
    
    def get_limit(self):
        return self._limit
    
    def get_balance(self):
        return self._balance
    
    def charge(self, price):
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True
    
    def make_payment(self, amount):
        if type(amount) != int or amount < 0:
            raise TypeError("Amount must be an integer and greater than 0")
        self._balance -= amount

    def _set_balance(self, balance):
        self._balance = balance
```

```python
class Vector:
    def __init__(self, d):
        self._coords = [0] * d if type(d) == int else d
    
    def __len__(self):
        return len(self._coords)
    
    def __getitem__(self, j):
        return self._coords[j]
    
    def __setitem__(self, j, val):
        self._coords[j] = val
    
    def __add__(self, other):
        if len(self) != len(other):
            raise ValueError("Dimensions must agree")
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] + other[j]
        return result
    
    def __radd__(self, other):
        return self + other
    
    def __eq__(self, other):
        return self._coords == other._coords
    
    def __ne__(self, other):
        return not self == other
    
    def __str__(self):
        return '<' + str(self._coords)[1:-1] + '>'
    
    def __sub__(self, other):
        if len(self) != len(other):
            raise ValueError("Dimensions must agree")
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] - other[j]
        return result
    
    def __neg__(self):
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = -self[j]
        return result
    
    def __mul__(self, other):
        if type(other) == int:
            result = Vector(len(self))
            for j in range(len(self)):
                result[j] = self[j] * other
            return result
        elif type(other) == Vector:
            if len(self) != len(other):
                raise ValueError("Dimensions must agree")
            result = 0
            for j in range(len(self)):
                result += self[j] * other[j]
            return result
        else:
            raise TypeError("Must be an integer or a Vector")
        
    
    def __rmul__(self, scalar):
        return self * scalar
        

    def __repr__(self) -> list[int]:
        return self._coords
```

```python
class Progression:
    def __init__(self, start=0):
        self._current = start
    
    def _advance(self):
        self._current += 1
    
    def __next__(self):
        if self._current is None:
            raise StopIteration()
        else:
            answer = self._current
            self._advance()
            return answer
    
    def __iter__(self):
        return self
    
    def print_progression(self, n):
        print(' '.join(str(next(self)) for j in range(n)))
```

```python
class DifferenceProgression(Progression):
    def __init__(self, num1=2, num2=200):
        self.prev = num2 - num1
        super().__init__(num1)

    def _advance(self):
        self.prev, self._current = self._current, abs(self._current - self.prev)


    
```

```python
iterator = DifferenceProgression()
iterator.print_progression(100)
```

```python
class RootProgression(Progression):
    def __init__(self, start=65536):
        super().__init__(start)
    
    def _advance(self):
        self._current = self._current ** 0.5

iterator = RootProgression()
iterator.print_progression(100)
```

```python
class FibonacciProgression(Progression):
    def __init__(self, first=0, second=1) -> None:
        super().__init__(first)
        self._prev = second - first
    
    def _advance(self):
        self._prev, self._current = self._current, self._prev + self._current
```

```python
fib_iterator = FibonacciProgression(2, 2)
for i in range(500):
    print(next(fib_iterator), end=' ')
```

```python
class ArithmeticProgression(Progression):
    def __init__(self, increment=1, start=0) -> None:
        super().__init__(start)
        self._increment = increment
    
    def _advance(self):
        self._current += self._increment
```

```python
class Sequence:
    def __init__(self, start=0):
        self._start = start
        self._changed = False
    
    def __getitem__(self, j):
        if self._changed:
            self._start = self._start + j
            self._changed = False
        else:
            self._changed = True
        return self._start
    
    def __eq__(self, other):
        return self._start == other._start
    
    def __lt__(self, other):
        return self._start < other._start
```

```python
class SequenceIterator:
    def __init__(self, sequence) -> None:
        self._seq = sequence
        self._k = -1
    
    def __next__(self):
        self._k += 1
        if self._k < len(self._seq):
            return self._seq[self._k]
        else:
            raise StopIteration()
    
    def __iter__(self):
        return self
```

```python
class ReversedSequenceIterator:
    def __init__(self, sequence) -> None:
        self._seq = sequence
        self._k = len(sequence)
    
    def __next__(self):
        self._k -= 1
        if self._k >= 0:
            return self._seq[self._k]
        else:
            raise StopIteration()
    
    def __iter__(self):
        return self
```

```python
class PredatoryCreditCard(CreditCard):
    LATE_FEE = 20

    def __init__(self, customer, bank, account, limit, apr) -> None:
        super().__init__(customer, bank, account, limit)
        self._apr = apr
        self._calls = 0
        self._minimum_monthly_payment = 0.05
        self.payed_thid_month = False
    
    def charge(self, price):
        success = super().charge(price)
        if not success:
            self._balance += 5
        return success
    
    def process_month(self):
        if not self.payed_thid_month:
            self._balance += self.LATE_FEE
        self._calls += 1
        if self._balance > 0:
            monthly_factor = pow(1 + self._apr, 1/12)
            self._balance *= monthly_factor
        if self._calls > 10:
            self._balance += 1
```
