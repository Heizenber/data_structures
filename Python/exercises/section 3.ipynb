```python
import matplotlib.pyplot as plt
import numpy as np
from math import log

# Define the range of values for n
n = np.linspace(1.1, 20, 100)

# Calculate the function values
f1 = 8 * n
f2 = 4 * n * np.log(n)
f3 = 2 * n**2
f4 = n**3
f5 = 2**n

# Plot the functions on a logarithmic scale
plt.plot(np.log(n), np.log(f1), label='8n')
plt.plot(np.log(n), np.log(f2), label='4nlogn')
plt.plot(np.log(n), np.log(f3), label='2n^2')
plt.plot(np.log(n), np.log(f4), label='n^3')
plt.plot(np.log(n), np.log(f5), label='2^n')

# Set the x-axis and y-axis labels
plt.xlabel('log(n)')
plt.ylabel('log(f(n))')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
n = np.linspace(1.1, 20, 100)

functions = [8 * n, 4 * n * np.log(n), 2 * n**2, n**3, 2**n]
labels = ['8n', '4nlogn', '2n^2', 'n^3', '2^n']

for i in range(len(functions)):
    plt.plot(np.log(n), np.log(functions[i]), label=labels[i])

plt.xlabel('log(n)')
plt.ylabel('log(f(n))')
plt.legend()
plt.show()
```
